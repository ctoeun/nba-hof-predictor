```python
# import necessary libraries
import requests
import urllib
from bs4 import BeautifulSoup
from urllib.request import urlopen

import pandas as pd
import scipy as sc
import numpy as np
import seaborn as sns
import re
import time
import os
import sys
import csv

import statsmodels.formula.api as sm

import matplotlib.pyplot as plt 
plt.style.use('ggplot')
%matplotlib inline  
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
years = list(range(1989,2023))

folders = ['csv_files', 'Webpages/Seasons', 'Webpages/Players'] 

for folder in folders:
    if not os.path.exists(folder):
            os.makedirs(folder)
            
for year in years:
    if not os.path.exists('Webpages/Seasons/{} Season'.format(year)):
        os.makedirs('Webpages/Seasons/{} Season'.format(year))
```

```python
# urls
award_years = 'https://www.basketball-reference.com/awards/awards_{}.html'
per_season_stats = 'https://www.basketball-reference.com/leagues/NBA_{}_per_game.html'
all_nba_league_teams = 'https://www.basketball-reference.com/awards/all_league.html'
team_records = 'https://www.basketball-reference.com/leagues/NBA_2021.html'

# directories
season_directory = 'Webpages/Seasons/{} Season{}'
per_season_stats_dir = 'Webpages/Players/Seasons/{}_season_player_stats.html'

for year in years[:-1]:
    urllib.request.urlretrieve(award_years.format(year), season_directory.format(year, '/{}_awards.html'.format(year)))
    urllib.request.urlretrieve(per_season_stats.format(year), per_season_stats_dir.format(year))
    urllib.request.urlretrieve(all_nba_league_teams, 'Webpages/all_awards.html')
    urllib.request.urlretrieve(team_records, season_directory.format(year, '/{} standings'.format(year)))
```

```python
dfs=[]
for year in years[:-1]:
    with open(awards_directory.format(year, year), 'rb') as html:
        nba_soup = BeautifulSoup(html)
    # remove headers
    for header in nba_soup.find_all('tr', class_='over_header'):
        header.decompose()
        
    nba_soup.find('tr', id='start_2nd').decompose()
    nba_soup.find('tr', id='start_3rd').decompose()
    nba_soup.find('tr', id='start_ORV').decompose()
    all_nba_table = nba_soup.find(id='leading_all_nba')
    all_nba_df = pd.read_html(str(all_nba_table))[0]
    all_nba_df['Year'] = year
    
    dfs.append(all_nba_df)
```

```python
all_nbas = pd.concat(dfs)
```

```python
all_nbas[['2nd Tm', '3rd Tm']] = all_nbas[['2nd Tm', '3rd Tm']].fillna(0)
all_nbas['Player'] = all_nbas['Player'].str.replace('*', '', regex=False)
```

```python
def merge_player_stats(df):
    if df.shape[0] == 1: return df
    else: row = df[df['Tm'] == 'TOT']; row['Tm'] = df.iloc[-1,:]['Tm']; return row

player_season_stats_csv = player_season_stats_csv.groupby(['Player', 'Year']).apply(merge_player_stats)
```

```python
dfs=[]

for year in years[:-1]:
    with open(season_directory.format(year, '/{} standings'.format(year)), 'rb') as html:
        nba_soup = BeautifulSoup(html)
    
    # remove headers
    for header in nba_soup.find_all('tr', class_='thead'):
        header.decompose()
    
    standings = nba_soup.find(id='divs_standings_E')
    standings_df = pd.read_html(str(standings))[0]
    standings_df['Year'] = year
    standings_df['Team'] = standings_df['Eastern Conference']
    del standings_df['Eastern Conference']
    dfs.append(standings_df)
        
    standings = nba_soup.find(id='divs_standings_W')
    standings_df = pd.read_html(str(standings))[0]
    standings_df['Year'] = year
    standings_df['Team'] = standings_df['Western Conference']
    del standings_df['Western Conference']
    dfs.append(standings_df)
```

```python
conference_standings_df = pd.concat(dfs)
```

```python
conference_standings_df['Team'] = conference_standings_df['Team'].str.replace('*', '', regex=False)
```

```python
conference_standings_df.insert(0, 'Team', conference_standings_df.pop('Team'))
```
