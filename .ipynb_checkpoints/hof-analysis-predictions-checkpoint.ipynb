## Introduction and Description

The *Naismith Memorial Basketball Hall of Fame* honors players who have shown exceptional skill at basketball, all-time great coaches, referees, and other major contributors to the sport. As part of the inaugural class of 1959, four players were inducted; over 150 more individuals have been inducted as players since then. As of 2022, players are eligible for the Hall of Fame after they have been fully retired from play for at least three years. 

A player or staff member can be considered a hall of famer for many different reasons; however, a predictive model is going to be trained on Hall of Fame eligible players to determine recently retired and current *NBA players* chances through stats, awards, and accomplishments. *If a player is currently active, we will assume they will retired at the end of the season.* <br>

#### Note:
**International Players:** There have been NBA players who have been known to have outstanding international careers, hence they were inducteed due to those acheivements. However, many of these player's NBA careers never really met expectations, so their stats may deviate from players who have strong accomplishments in the NBA. Regardless, I won't remove these players so there is no bias. <br>

**Coaches and Contributors:** Many Hall of Fame coaches such as Phil Jackson, Pat Riley, and Jerry Sloan have had careers as players, however, most were average. In addition, a few Hall of Fame players have been inducted as contributors of the NBA. These include game analysts, reporters, journalist, and many others. When computing the correlation matrix, I won't be counting these players to remove outliers (Hall of Fame players will be labled based on how they were inducted: Player, Coach, or Contributor). <br>

```python
fig, axs = plt.subplots(1,2,figsize=(25, 11))

# per game stats graph
graph_hof_basic = retired_hof_eligible.groupby('Inducted as Player')['Career PPG','Career RPG',
                                                           'Career APG','Career SPG','Career BPG']
graph_hof_basic = graph_hof_basic.mean().transpose()
graph_hof_basic = graph_hof_basic.rename({'Career PPG': 'PTS', 
                                  'Career RPG': 'RBS', 
                                  'Career APG': 'AST',
                                   'Career SPG': 'STL',
                                    'Career BPG': 'BLK'},
                                   axis='index')
axs[0].xaxis.set_tick_params(labelsize=25); axs[0].yaxis.set_tick_params(labelsize=20)
graph_hof_basic = graph_hof_basic.rename({1:'HOF', 0:'NON-HOF'}, axis='columns')
graph_hof_basic[graph_hof_basic.columns[::-1]].plot(kind='bar',color={'HOF':'Blue','NON-HOF':'Red'},
                                                    rot=0, alpha=0.6, ax=axs[0])
axs[0].set_title('Average Per Game Stats by Hall of Fame Status', fontsize=25, fontweight="bold")
axs[0].legend(loc=1, frameon=True, prop={'size': 20}).remove()

# percentage graphs
graph_hof_percentage = retired_hof_eligible.groupby('Inducted as Player')['Career FG%','Career FG3%',
                                                            'Career FT%','Career eFG%', 'PER', 'Avg WS']
graph_hof_percentage = graph_hof_percentage.mean().transpose()
graph_hof_percentage = graph_hof_percentage.rename({'Career FG%': 'FG%', 
                                  'Career FG3%': 'FG3%', 
                                  'Career FT%': 'FT%',
                                   'Career eFG%': 'eFG%'},
                                   axis='index')
axs[1].xaxis.set_tick_params(labelsize=25); axs[1].yaxis.set_tick_params(labelsize=20)
graph_hof_percentage = graph_hof_percentage.rename({1:'HOF', 0:'NON-HOF'}, axis='columns')
graph_hof_percentage[graph_hof_percentage.columns[::-1]].plot(kind='bar',color={'HOF':'Blue','NON-HOF':'Red'},
                                                rot=0, alpha=0.6, ax=axs[1])

axs[1].set_title('Average Per Game Percentages by Hall of Fame Status', fontsize=25, fontweight="bold")
axs[1].legend(loc=1, frameon=True, prop={'size': 20})

fig.tight_layout(pad=3.5)

plt.savefig('graphs/bar_graphs_averages.png')
plt.show()
```

Like I have stated above, the Per Game stats are not surprising. All Hall of Fame players average about *17ppg, 8rpg, 4apg, 1spg, and under 1bpg*. The bar plot emphasises that this more than double the average of retired Non-Hall of Famers.

In terms of percentages, the differences are less drastic. With the current Hall of Fame class being center dominate, Field Goal percentages and Three Point percentages are quite similar. However, Efficiency Field Goal is a bit higher for Non-Hall of Famers. I would infer this is because there are a decent amount of centers who have a low Three Point percentage.

### Scatter Plots:

```python
colors = ['#FFC125','#CD0000','#3A5FCD'] # gold, red, blue

def get_color(df):
    if df['Inducted as Player'] == 1:
        if df['First Ballot'] == 1: return colors[0]
        else: return colors[1]
    else: return colors[2]
```

```python
fig, axs = plt.subplots(3,figsize=(13,25))

# total games and points
retired_hof_eligible.plot.scatter('Total PTS', 'Total Games', 
                    c=retired_hof_eligible.apply(lambda row: get_color(row), axis=1), colormap='jet', ax=axs[0])
axs[0].set_title('Total Games and Points by Hall of Fame Status', fontsize=20, fontweight="bold")
markers = [plt.Line2D([0,0],[0,0],color=color, marker='o', linestyle='') for color in colors]
axs[0].legend(markers, ['First Ballot','HOF (Non-First Ballot)','NON-HOF'], prop={'size':12}, numpoints=1, loc='lower right')

# total rebounds and assists
retired_hof_eligible.plot.scatter('Total RBS', 'Total AST', 
                    c=retired_hof_eligible.apply(lambda row: get_color(row), axis=1), colormap='jet', ax=axs[1])

axs[1].set_title('Total Rebounds and Assists by Hall of Fame Status', fontsize=20, fontweight="bold")
markers = [plt.Line2D([0,0],[0,0],color=color, marker='o', linestyle='') for color in colors]
axs[1].legend(markers, ['First Ballot','HOF (Non-First Ballot)','NON-HOF'], prop={'size':12}, numpoints=1, loc='lower right')

# total steals and blocks
retired_hof_eligible.plot.scatter('Total BLK', 'Total STL', 
                    c=retired_hof_eligible.apply(lambda row: get_color(row), axis=1), colormap='jet', ax=axs[2])

axs[2].set_title('Total Blocks and Steals by Hall of Fame Status', fontsize=20, fontweight="bold")
markers = [plt.Line2D([0,0],[0,0],color=color, marker='o', linestyle='') for color in colors]
axs[2].legend(markers, ['First Ballot','HOF (Non-First Ballot)','NON-HOF'], prop={'size':12}, numpoints=1, loc='lower right')

fig.tight_layout(pad=3.0)

plt.savefig('graphs/scatter_plots.png')
plt.show()
```

Those who score more than 20,000 career points are practically guaranteed to be inducted into the Hall of Fame as a player and likely as a First Ballot. The same applies with Rebounds, Assists, Steals, and Blocks. <br>

A player is guaranteed a spot with these total stats (based on current Hall of Famers):
- Games: > 1,500
- Points: > 20,000
- Rebounds: > 13,000
- Assists: > 11,000
- Steals: > 2,200
- Blocks: > 3,200

```python
fig, axs = plt.subplots(3,figsize=(13,25))

hof_type = {1:'tab:red',0:'tab:blue'}
# championships
retired_hof_eligible.plot.scatter('NBA Champs', 'Total Games', 
                    c=retired_hof_eligible.apply(lambda row: get_color(row), axis=1), colormap='jet',
                                 ax=axs[0])

axs[0].set_title('NBA Championships by Hall of Fame Status', fontsize=20, fontweight="bold")
markers = [plt.Line2D([0,0],[0,0],color=color, marker='o', linestyle='') for color in colors]
axs[0].legend(markers, ['First Ballot','HOF (Non-First Ballot)','NON-HOF'], prop={'size':12}, numpoints=1, loc='lower right')

# mvps
retired_hof_eligible.plot.scatter('MVPs', 'Total Games', 
                    c=retired_hof_eligible.apply(lambda row: get_color(row), axis=1), colormap='jet',
                                 ax=axs[1])

axs[1].set_title('MVPs by Hall of Fame Status', fontsize=20, fontweight="bold")
markers = [plt.Line2D([0,0],[0,0],color=color, marker='o', linestyle='') for color in colors]
axs[1].legend(markers, ['First Ballot','HOF (Non-First Ballot)','NON-HOF'], prop={'size':12}, numpoints=1, loc='lower right')

# finals mvps
retired_hof_eligible.plot.scatter('Finals MVPs', 'Total Games', 
                    c=retired_hof_eligible.apply(lambda row: get_color(row), axis=1), colormap='jet',
                                 ax=axs[2])

axs[2].set_title('Finals MVPs by Hall of Fame Status', fontsize=20, fontweight="bold")
markers = [plt.Line2D([0,0],[0,0],color=color, marker='o', linestyle='') for color in colors]
axs[2].legend(markers, ['First Ballot','HOF (Non-First Ballot)','NON-HOF'], prop={'size':12}, numpoints=1, loc='lower right')

fig.tight_layout(pad=3.5)

plt.savefig('graphs/champs_and_mvps.png')
plt.show()
```

Most consider NBA championships, MVPs, and Finals MVPs to be strong factors in determining a players chance of being a Hall of Fame. However, these scatter plots show that is not the case. However, if a player were to have at least one of each, their chances increase quite dramatically. <br>

The first scatter plot shows that the number of NBA Championships isn't a huge factor in terms of deciding Hall of Fame status. Many who are first ballot hadn't even won one and many who have won a few haven't been inducted. Championships are a team achievement and only a few players on a championship team have strong contributions. So bench players on a good team can still win. <br>

**For example, let's take a look at Patrick McCaw:**
![pmccaw.png](images/pmccaw.png) <br>
Patrick McCaw is a player with 3 NBA Championships (Two with GSW and One with TOR). However, he averaged *4ppg, 2rpg, and 1apg*. Comparing his individual stats with Michael Jordan, or any other player, many would consider him as a below average player and non-Hall of Famer. Golden State and Toronto were the top two teams during the year and had superstars such as Steph Curry/Kevin Durant (GSW) and Kyle Lowry/Kawhi Leonard (TOR). These facts conclude the that the number of championships alone don't determine a players greatness and increase chances of becoming a Hall of Famer. There are many other factors that need to be included. <br>

Steve Nash is a particular First Ballot Hall of Famer with no championships:
![nash.png](images/nash.png) <br>

The differences are quite large when comparing the two players. Steve Nash has other seasonal awards and great per game averages that strengthen his resume. Although a player with few championships combined with outstanding per game stats and seasonal awards has a high chance of being a First Ballot Hall of Famer. <br>

In terms of number of MVPs and Finals MVPs, both arguments could be the same. If a player were to win at least one MVP or two Final MVPs, the scatter plots show they are guaranteed a Hall of Fame spot (granted they have other stats and awards to complement). It is hard to see, but there are a couple players with one Final MVP who are still aren't inducted as a player.

## Analysis Methodology: Predictive Analysis

Lets go ahead and do some predictions. These are the main goals:
- Use a probabilistic classfication model to predict top players for the 2023 Hall of Fame class and predict the probability of future induction for active players. 
- Use a binary classification model to predict the whether a recently retired and active player is a First Ballot Hall of Famer. <br>

These are the models I will choose from based on accuracy:
- Probabilistic Classification:
    - Linear Regression
    - SVM (Using platt scaling)
    - Naive Bayes Classifer
- Binary Classification:
    - SVM (Support Vector Machine)
    - Decision Tree
    - Random Forest Classifer
    - K-NN (K-Nearest Neighbors)

The top models for each classification will be used for predictions.

We'll use X, Y to predict probability for being inducted as a player and x,y for predicting whether a player is a first ballot hall of famer (using binary).

```python
X, Y = retired_hof_eligible[hof_predictive_variables], retired_hof_eligible['Inducted as Player']
x, y = retired_hof_eligible[fst_ballot_pred_var], retired_hof_eligible['First Ballot']
```

The sample size is quite small, so there isn't much need to do a train-test split.

```python
# X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

First lets train each model on retired players eligible to be inducted (this includes Hall of Fame inductees as well) and find which provides the best accuracy in terms of precision and F1-score.

### Probabilistic Classification:

#### Logistic Regression

```python
logistic_reg = LogisticRegression(solver='liblinear', C=4.0, random_state=42).fit(X, Y)
y_pred = logistic_reg.predict(X)

print(metrics.classification_report(Y, y_pred)); print()
print('Accuracy:', metrics.accuracy_score(Y, y_pred))
print('Mean Abs Error: ', metrics.mean_absolute_error(Y, y_pred))
print('Mean Squared Error: ', metrics.mean_squared_error(Y, y_pred)); print()
print(metrics.confusion_matrix(Y, y_pred)); print()
```

#### Support Vector Machine (SVM): Using Platt scaling

```python
# use platt scaling to calibrate and transform into probablity model
svm_model = svm.SVC(kernel='linear', random_state=42, probability=True).fit(X, Y)
y_pred = svm_model.predict(X)

print(metrics.classification_report(Y, y_pred)); print()
print('Accuracy:', metrics.accuracy_score(Y, y_pred))
print('Mean Abs Error: ', metrics.mean_absolute_error(Y, y_pred))
print('Mean Squared Error: ', metrics.mean_squared_error(Y, y_pred)); print()
print(metrics.confusion_matrix(Y, y_pred)); print()
```

#### Gaussian Naive Bayes

```python
bayes = GaussianNB().fit(X, Y)
y_pred = bayes.predict(X)

print(metrics.classification_report(Y, y_pred)); print()
print('Accuracy: ', metrics.accuracy_score(Y, y_pred))
print('Mean Abs Error: ', metrics.mean_absolute_error(Y, y_pred))
print('Mean Squared Error: ', metrics.mean_squared_error(Y, y_pred)); print()
print(metrics.confusion_matrix(Y, y_pred)); print()
```

### Binary Classification:

#### Support Vector Machine (SVM)

```python
svm_binary = svm.SVC(random_state=42).fit(x, y)
y_pred = svm_binary.predict(x)

print(classification_report(y, y_pred)); print()
print('Accuracy: ', accuracy_score(y, y_pred))
print('Mean Abs Error: ', metrics.mean_absolute_error(y, y_pred))
print('Mean Squared Error: ', metrics.mean_squared_error(y, y_pred)); print()
print(confusion_matrix(y, y_pred)); print()
```

#### Decision Tree

```python
decision_tree = tree.DecisionTreeClassifier(max_depth=5, random_state=42).fit(x, y)
y_pred = decision_tree.predict(x)

print(classification_report(y, y_pred)); print()
print('Accuracy: ', accuracy_score(y, y_pred))
print('Mean Abs Error: ', metrics.mean_absolute_error(y, y_pred))
print('Mean Squared Error: ', metrics.mean_squared_error(y, y_pred)); print()
print(confusion_matrix(y, y_pred)); print()
```

#### Random Forest Classifer

```python
rand_forest = RandomForestClassifier(n_estimators=500, random_state=42, max_depth=10).fit(x, y)
y_pred = rand_forest.predict(x)

print(classification_report(y, y_pred)); print()
print('Accuracy: ', accuracy_score(y, y_pred))
print('Mean Abs Error: ', metrics.mean_absolute_error(y, y))
print('Mean Squared Error: ', metrics.mean_squared_error(y, y_pred)); print()
print(confusion_matrix(y, y_pred)); print()
```

#### K-Nearest Neighbors (K-NN)

```python
knn = KNeighborsClassifier(n_neighbors=10).fit(x, y)
y_pred = knn.predict(x)

print(classification_report(y, y_pred)); print()
print('Accuracy: ', accuracy_score(y, y_pred))
print('Mean Abs Error: ', metrics.mean_absolute_error(y, y_pred))
print('Mean Squared Error: ', metrics.mean_squared_error(y, y_pred)); print()
print(confusion_matrix(y, y_pred)); print()
```

After training each model, we can conclude **Linear Regression** (Probability) and **Random Forest Classifer** (Binary) provide the best accuracies, so we'll use them in the predictions.

```python
# function for predicting hof probability for all nba players
def hof_prob(row):
    player = row[hof_predictive_variables]
    hof_probability = logistic_reg.predict_proba([player]);
    return hof_probability[0][1]
```

```python
# function for predicting whether a player should be a first ballot hofer (binary)
def first_ballot(row):
    player = row[fst_ballot_pred_var]
    hof_binary = rand_forest.predict([player]);
    return hof_binary[0]
```
